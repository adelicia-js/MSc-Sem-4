# Image Processing - Assignment 1

```python
# Import libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Task 1 - Open & Display an image.

```python
imagePath = './images/hammie.jpg'
hamster = cv2.imread(imagePath)
imageRGB = cv2.cvtColor(hamster, cv2.COLOR_BGR2RGB)

plt.imshow(imageRGB)
plt.title('Cute baby hamster')
plt.axis('off')

plt.show()
```

### Task 2 - Display the intensity value at \(10,10\).

```python
intensityVal = hamster[10,10]
print("Intensity value at (10,10): ", intensityVal)
```

### Task 3 - Convert an image to grayscale and b/w.

```python
grayImg = cv2.cvtColor(hamster, cv2.COLOR_BGR2GRAY)

plt.imshow(grayImg, cmap='gray')
plt.title('Grayscale hamster')
plt.axis('off')
plt.show()
```

```python
ret, bwImg = cv2.threshold(grayImg, 127, 255, cv2.THRESH_BINARY)

plt.imshow(bwImg, cmap='binary')
plt.title ('B/w hamster')
plt.axis('off')
plt.show()
```

### Task 4 - Extract the difference between two similar images.

```python
horseOne = cv2.imread('./images/horse1.png')

plt.imshow(cv2.cvtColor(horseOne, cv2.COLOR_BGR2RGB))
plt.title('Horse One')
plt.axis('off')
plt.show()
```

```python
horseTwo = cv2.imread('./images/horse2.png')

plt.imshow(cv2.cvtColor(horseTwo, cv2.COLOR_BGR2RGB))
plt.title('Horse Two')
plt.axis('off')
plt.show()
```

```python
# Check shapes
print(horseOne.shape, horseTwo.shape)
```

```python
# Resize shapes conditionally
if horseOne.shape[0] < horseTwo.shape[0] and horseOne.shape[1] < horseTwo.shape[1]:
    horseOne = cv2.resize(horseOne, (horseTwo.shape[1], horseTwo.shape[0]))
elif horseTwo.shape[0] < horseOne.shape[0] and horseTwo.shape[1] < horseOne.shape[1]:
    horseTwo = cv2.resize(horseTwo, (horseOne.shape[1], horseOne.shape[0]))

print(horseOne.shape, horseTwo.shape)
```

```python
difference = cv2.absdiff(horseOne, horseTwo)

plt.imshow(cv2.cvtColor(difference, cv2.COLOR_BGR2RGB))
plt.title('Differences between the two images')
plt.axis('off')
plt.show()
```

The differences are the tail shape & the horse's smile.

### Task 5 - Apply all logical operations on an image.

```python
backdrop = cv2.imread('./images/bg.jpg')

plt.imshow(cv2.cvtColor(backdrop, cv2.COLOR_BGR2RGB))
plt.title('A scene')
plt.axis('off')
plt.show()
```

```python
woman = cv2.imread('./images/woman.jpg')

plt.imshow(cv2.cvtColor(woman, cv2.COLOR_BGR2RGB))
plt.title('A woman playing piano')
plt.axis('off')
plt.show()
```

```python
# Check shapes of both
print(woman.shape, backdrop.shape)
```

```python
# Resize shapes
backdrop = cv2.resize(backdrop, (woman.shape[1], woman.shape[0]))
print(woman.shape, backdrop.shape)
```

#### Logical AND Operation

```python
logicalAnd = cv2.bitwise_and(woman, backdrop)

plt.imshow(logicalAnd, cmap='gray')
plt.title('Logical AND')
plt.axis('off')
plt.show()
```

#### Logical OR Operation

```python
logicalOr = cv2.bitwise_or(woman, backdrop)

plt.imshow(logicalOr, cmap='gray')
plt.title('Logical OR')
plt.axis('off')
plt.show()
```

#### Logical NOT Operation

```python
logicalNot = cv2.bitwise_not(woman)

plt.imshow(logicalNot, cmap='gray')
plt.title('Logical NOT')
plt.axis('off')
plt.show()
```

#### Logical AND-NOT Operation

```python
logicalAndNot = cv2.bitwise_and(woman, cv2.bitwise_not(backdrop))

plt.imshow(logicalAndNot, cmap='gray')
plt.title('Logical AND-NOT')
plt.axis('off')
plt.show()
```

#### Logical XOR Operation

```python
logicalXor = cv2.bitwise_xor(woman, backdrop)

plt.imshow(logicalXor, cmap='gray')
plt.title('Logical XOR')
plt.axis('off')
plt.show()
```
